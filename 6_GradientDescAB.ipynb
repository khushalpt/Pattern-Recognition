```python
#importing necessary libraries

import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
import decimal
decimal.getcontext().prec = 100

#Actual Requested function 6_a
def proposed_functionA(b):
     return b*b
    
def derivA(b):
    b_deriv = 2*b
    return b_deriv

#Derivation using gradient desc for 6_a

def GradientDesc_A(b_new,b_old,learning,precision_scale):
   
   
    x_list, y_list = [b_new], [proposed_functionA(b_new)]
    while abs(b_new-b_old) > precision_scale:
        b_old = b_new
        d_x = - derivA(b_old)
        b_new = b_old + (learning * d_x)
        x_list.append(b_new)
        y_list.append(proposed_functionA(b_new))
    print ("Local minimum occurs at: "+ str(b_new))
    print ("Number of steps taken: " + str(len(x_list)))
   
    plt.subplot(1,2,2)
    plt.scatter(x_list,y_list,c="g")
    plt.plot(x_list,y_list,c="g")
    x=np.linspace(-1,3,500)
    plt.plot(x,proposed_functionA(x), c="r")
    plt.title("Gradient descent Graph for 6a")
    plt.show()

#Actual Requested function 6_b
def proposed_functionB(b_new):
    return ((math.sin(10 * 3.14 * b_new) / 2 *b_new) + ((b_new-1)**4))
def derivB(b):
    b_deriv = ((((10* 3.14* math.cos(10 * 3.14* b))-(math.sin(10*3.14*b)))/2*(b**2)) + 4 * ((b-1)**3))
    return b_deriv

#Derivation using gradient desc for 6_b
def GradientDesc_B(b_new,b_old,learning,precision_scale):

    x_list, y_list = [b_new], [proposed_functionB(b_new)]
    while (b_new-b_old) > precision_scale:
        b_old = b_new
        d_x = -derivB(b_new)
        b_new = b_old + (learning * d_x)
        x_list.append(b_new)
        
        y_list.append(proposed_functionB(b_new))
        

    print ("Local minimum occurs at: "+ str(b_new))
    print ("Number of steps taken: " + str(len(x_list)))
   
   
    plt.subplot(1,2,2)
    plt.scatter(x_list,y_list,c="g")
    plt.plot(x_list,y_list,c="g")
    x = np.linspace(-3,3,500)
    func_vector=np.vectorize(proposed_functionB)
    plt.plot(x,func_vector(x), c="r")
    plt.title("Gradient descent graph for 6b")
    plt.show()
```
