```python
#importing necessary libraries

from os import listdir
from os.path import isfile, join
from random import shuffle
import pandas as pd
import numpy as np
import math
import operator
    
def euclidean(obj1, obj2, dis):  #function to calculate euclidean measure
    dist = 0
    for x in range(dis):
        for i in range(x):
            dist+= pow((obj1[x] - obj2[x]),2) #applying the distance formula
            y = math.sqrt(dist)
    
    return math.sqrt(dist)

def getNeigh(train_set, test_obj, k): #function to have the neighbours calculated
    distances = []
    dis = len(test_obj)-1
       
    for x in range(len(train_set)):
        dist=euclidean(test_obj, train_set[x], dis)
        distances.append((train_set[x],dist))
    distances.sort(key=operator.itemgetter(1))
    neighbours = []
    #for i in range(k):
    for x in range(k):
        #neighbours.append(distances[i][0])
        neighbours.append(distances[x][0])
    return neighbours


def getResp(neighbours):
    class_comp={}
    for x in range(len(neighbours)):
        resp = neighbours[x][-1]
        if resp in class_comp:
            class_comp[resp] += 1
        else:
            class_comp[resp] = 1
    sort_comp = sorted(class_comp.items(),key = operator.itemgetter(1), reverse=True)
    return sort_comp[0][0]

def getAcc(tests, predictions):   #Provide the Accuracy measure of test and predicted data
    correct = 0
    for x in range(len(tests)):
        if tests[x][-1] == predictions[x]:
            correct += 1
           
    return (correct/float(len(tests))) * 100


def knn():
    

    myPath =r'C:\Users\khushalpt\training_validation' #openiing files

    data_files = listdir(myPath)
    print("Number of Files: " + str(len(data_files)))
    classificationSet = []
    datas = []

    for fileName in data_files :  
        with open(join (myPath,fileName)) as file:                      
            datas.append(list(file.read().replace('\n', '')))
            classificationSet.append(int(str(fileName.split('_')[1]))) # appending label at initial position
    for i in range(0,len(datas)):
        datas[i]= list(map(int,datas[i])) #mapping data as integers
   
    dataset = [list(a) for a in zip(datas, classificationSet)]  


    def splitting_data(dataset,training=[],test=[]):
        shuffle(dataset)
        a = (int(len(dataset)/5))
        b = (int(a*(5-1)))
        training.extend(dataset[:b])
        test.extend(dataset[b:])
#output
    training=[]
    test=[]
    splitting_data(dataset,training,test)
    
    print("training Measure: "+ repr(len(training)))   #print(repr(training))
    print("test Measure: "+ repr(len(test)))
    predict=[]
    k=11

    for y in range(len(test)):
        neigh_list = getNeigh(training,test[y],k)
        r_list = getResp(neigh_list)
        predict.append(r_list)
    print('predicted=' + repr(r_list) + ', actual=' + repr(test[y][-1]))
    print(len(predict))
    acc = getAcc(test, predict)
    print('acc: ' + repr(acc) + '%')
   
knn()

```
